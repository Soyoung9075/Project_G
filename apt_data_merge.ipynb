## 분양가 리스트 파일 기준으로 나머지 파일 merge 하기

### 1. 분양가 리스트와 아파트 리스트(지역, 주택구분, 시공사) merge 하기

```python
# merge에 기준이 되는 분양가 리스트 데이터 shape 확인
price_df.shape
```

```python
# apt_df 에서 '지역', '주택구분' '분양/임대', '시공사' merge

selected_columns = ['아파트이름', '지역', '주택구분', '분양/임대', '시공사']
merge_1 = pd.merge(price_df, apt_df[selected_columns], how = 'left', on = '아파트이름')
merge_1
```

=> price_df 기준으로 merge 하면 rows 수가 10643이어야 하는데 10802임

=> apt_df에 중복 있는지 확인 필요

```python
# apt_df에 중복 있음. 
duplicated_rows = apt_df[apt_df['아파트이름'].duplicated(keep=False)]['아파트이름']
print(duplicated_rows)
```

=> 확인결과 입주자 모집을 두 번 공고낸 케이스들.

=> 일단 삭제

```python
# apt_df의 아파트명 중복 되는 것 삭제

apt_df.drop_duplicates(subset=['아파트이름'], ignore_index = True, keep= False, inplace = True)
```

```python
# 중복 있는지 다시 확인
duplicated_rows_2 = apt_df[apt_df['아파트이름'].duplicated(keep=False)]['아파트이름']
print(duplicated_rows_2)
# => 없음
```

```python
# price_df에도 해당 아파트 명 삭제  (총 122 row 삭제)

duplicated_list = list(set(duplicated_rows))

price_df_drop = price_df.drop(price_df.index[price_df['아파트이름'].isin(duplicated_list)]).reset_index(drop = True)
price_df_drop.shape
```

```python
# 다시 merge하기
selected_columns = ['아파트이름', '지역', '주택구분', '분양/임대', '시공사']
merge_1 = pd.merge(price_df_drop, apt_df[selected_columns], how = 'left', on = '아파트이름')
merge_1
```

=> 지역, 주택구분, 분양/임대, 시공사가 누락된 고덕센트럴과 센텀 SK VIEW는 수기로 채워넣기

```python
# 누락된 정보 수기로 채워넣기

# 고덕신도시 EBC-2BL 힐스테이트 고덕 센트럴
# 시공사 : 현대엔지니어링 주식회사
# 지역 : 경기
# 주택구분 : 민영
# 분양/임대 : 분양주택

# 센텀 아스트룸 SK VIEW
# 시공사 : SK에코플랜트(주)
# 지역 : 부산
# 주택구분 : 민영
# 분양/임대 : 분양주택

merge_2.loc[merge_2['아파트이름'] == '고덕신도시 EBC-2BL 힐스테이트 고덕 센트럴', '지역'] = '경기'
merge_2.loc[merge_2['아파트이름'] == '고덕신도시 EBC-2BL 힐스테이트 고덕 센트럴', '주택구분'] = '민영'
merge_2.loc[merge_2['아파트이름'] == '고덕신도시 EBC-2BL 힐스테이트 고덕 센트럴', '분양/임대'] = '분양주택'
merge_2.loc[merge_2['아파트이름'] == '고덕신도시 EBC-2BL 힐스테이트 고덕 센트럴', '시공사'] = '현대엔지니어링 주식회사'

merge_2.loc[merge_2['아파트이름'] == '센텀 아스트룸 SK VIEW', '지역'] = '부산'
merge_2.loc[merge_2['아파트이름'] == '센텀 아스트룸 SK VIEW', '주택구분'] = '민영'
merge_2.loc[merge_2['아파트이름'] == '센텀 아스트룸 SK VIEW', '분양/임대'] = '분양주택'
merge_2.loc[merge_2['아파트이름'] == '센텀 아스트룸 SK VIEW', '시공사'] = 'SK에코플랜트(주)'
```
